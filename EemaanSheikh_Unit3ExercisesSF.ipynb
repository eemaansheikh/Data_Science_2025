## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

### Playing with Beta

**Task1**:

Plot a [preliZ Beta distribution](https://preliz.readthedocs.io/en/latest/api_reference.html#preliz.distributions.beta.Beta) with $α=1$ and $β=1$.



````python

from preliz import Beta, style
style.use('preliz-doc')
alphas = [1]
betas = [1]
for alpha, beta in zip(alphas, betas):
    ax = Beta(alpha, beta).plot_pdf()
ax.set_ylim(0, 5)
````

**Task4**:

Then, plot these on the same graph: $Exponential(λ=1/2)$, $Gamma(α=2,β=1)$, $HalfStudentT(ν=5,σ1)$

For the t distribution and half t, use

```

```

respectively, inside the plot_pdf function.


````python
from preliz import Normal, style
style.use('preliz-doc')
mus = [0]
sigmas = [1]
for mu, sigma in zip(mus, sigmas):
    Normal(mu, sigma).plot_pdf()

from preliz import StudentT, style
style.use('preliz-doc')
nus = [5]
mus = [0]
sigmas = [1]
for nu, mu, sigma in zip(nus, mus, sigmas):
    StudentT(nu, mu, sigma).plot_pdf(support=(-4,4))
````

````python
from preliz import Exponential, style
style.use('preliz-doc')
for lam in [0.5]:
    Exponential(lam).plot_pdf(support=(0,5))

from preliz import Gamma, style
style.use('preliz-doc')
alphas = [2]
betas = [1]
for alpha, beta in zip(alphas, betas):
    Gamma(alpha, beta).plot_pdf()

from preliz import HalfStudentT, style
style.use('preliz-doc')
sigmas = [1]
nus = [5]
for sigma, nu in zip(sigmas, nus):
    HalfStudentT(nu, sigma).plot_pdf(support=(0,4))
````

###Playing with Discrete Distributions

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(μ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

Additionally, compute the first four moments for all those distributions--except the Categorical distribution.

````python
from preliz import Bernoulli, style
style.use('preliz-doc')
for p in [0.7]:
    Bernoulli(p).plot_pdf()


from preliz import Binomial, style
style.use('preliz-doc')
ns = [10]
ps = [0.7]
for n, p in zip(ns, ps):
    Binomial(n, p).plot_pdf()

from preliz import Poisson, style
style.use('preliz-doc')
for mu in [7]:
    Poisson(mu).plot_pdf()

from preliz import Geometric, style
style.use('preliz-doc')
for p in [0.7]:
    Geometric(p).plot_pdf(support=(1,10))

from preliz import HyperGeometric, style
style.use('preliz-doc')
N = 50
k = 35
for n in [15]:
    HyperGeometric(N, k, n).plot_pdf(support=(1,15))

from preliz import Categorical, style
style.use('preliz-doc')
ps = [0.05, 0.3, 0.05, 0.1, 0.2, 0.4 ]
for p in ps:
    Categorical(p).plot_pdf()
````
